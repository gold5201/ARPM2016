```python
import numpy as np 
import pandas as pd
import pandas_datareader.data as web
import datetime as dt
```

```python
SP500 = pd.read_csv('./SPY_All_Holdings.csv')
SP500['Weight'] = SP500['Weight']/100
SP500
```

```python
prices = pd.DataFrame()
start = dt.datetime(2016,1,1)
tickers = SP500['Ticker']
for ticker in tickers:
    prices[ticker] = web.DataReader(ticker, 'yahoo', start)['Adj Close']
returns = prices.pct_change()
returns
```

```python
compound_returns = returns.compound()
compound_returns = compound_returns.to_frame()
compound_returns.columns = ['Compound Return']
compound_returns['Ticker'] = compound_returns.index
compound_returns = compound_returns.reset_index(drop=True)
compound_returns
```

```python
SP500 = pd.merge(SP500, compound_returns, on='Ticker')
SP500 
```

```python
sectors = SP500.groupby('Sector')
sectors = sectors.sum()
sectors
```
